````python
#LRU 
def solution(n,pages):
    chair = []
    time = 0
    for p in pages:
        if not p in chair:
            if len(chair) < n:
                chair.append(p)
            else:
                chair.pop(0)
                chair.append(p)
            time +=60
        else:
            chair.append(chair.pop(chair.index(p)))
            time+=1
        print(chair)
    return time

pages = ['', '', '', '', '', '', '', '']
time = solution(3,pages)
print('{} {}'.format(time//60, time%60))
        
        
````
